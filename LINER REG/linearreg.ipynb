```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df = pd.read_csv("csv-delimeter.csv")
df
```

```python
x = df[['x']]
y = df[['y']]
```

```python
sns.scatterplot(df,x="x",y="y")
```

```python
sns.regplot(df,x="x",y="y")
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
print(lr.predict([[18]]))
```

```python
y_pred = lr.predict(x)
```

```python
print(mean_absolute_error(y,y_pred))#(y-y_pred/total n)
print(mean_squared_error(y,y_pred))#(y-y_pred/total n)2
print(r2_score(y,y_pred)) # accuracy of  model
```
