```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("csv-delimeter.csv")
# df = pd.DataFrame(data)
data
```

```python
sns.scatterplot(data,x ='x',y ='y')
```

```python
sns.regplot(data,x ='x',y ='y')
```

```python
# differecnr known as residule or residule error between line and point
```

```python
# coefficent as m or slop
# y is know as intersect

# m = coeffient
# c = constant / intersect
```

```python
# x = data['x']
# y = data['y']
x = data[['x']]
y = data[['y']]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
print(lr.predict([[18]]))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(x)
```

```python
print(mean_absolute_error(y,y_pred))#(y-y_pred/total n)
print(mean_squared_error(y,y_pred))#(y-y_pred/total n)2
print(r2_score(y,y_pred)) # accuracy of  model
```
