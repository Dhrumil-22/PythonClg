## MERGE

```python
import pandas as pd

df1 = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['a','b','c','d'],
    'dept_id':[10,20,30,40]
})
df2 = pd.DataFrame({
    'dept_id':[10,20,30],
    'dept' : ['IT','HR','FINANCE']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on='dept_id',how="inner")
```

```python
pd.merge(df1,df2,on='dept_id',how="left")
```

```python
pd.merge(df1,df2,on='dept_id',how="outer")
```

```python
pd.merge(df1,df2,on='dept_id',how="right")
```

```python


df1 = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['a','b','c','d'],
    'dept_id':[10,20,30,40]
})
df2 = pd.DataFrame({
    'd_id':[10,20,30],
    'dept' : ['IT','HR','FINANCE']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on="d_id")
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how="outer")
```

```python


df1 = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['a','b','c','d'],
    'dept_id':[10,20,10,40]
})
df2 = pd.DataFrame({
    'd_id':[10,20,30],
    'name' : ['IT','HR','FINANCE']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how="outer",suffixes=['_emp','_dept'])
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='d_id')
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='d_id',sort=False)
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='d_id',sort=True)
```
