## CONCAT

```python
df1 = pd.DataFrame({'A':[1,2],'B':[6,7]})
df2 = pd.DataFrame({'A':[3,4],'B':[9,5]})

print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1 = pd.DataFrame({'A':[1,2],'B':[6,7]})
df2 = pd.DataFrame({'A':[3,4],'D':[9,5]})
print(df1)
print(df2)
```

inner ma command colum hase ae rese
outer ma badhi column lese

default outer che

```python
pd.concat([df1,df2],join='outer')
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2],join='inner',ignore_index=True)
```
