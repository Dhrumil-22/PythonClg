```python
q1 = df["acceleration"].quantile(0.25)
q3 = df["acceleration"].quantile(0.75)
iqr = q3-q1
ur = q3+1.5*iqr
lr = q1 - 1.5*iqr
df = df.loc[(df["acceleration"]>lr) & (df["acceleration"]<ur)]
```

```python
df.info()
```

```python
sns.boxplot(data=df['cylinders'])
plt.show()
```

```python
q1 = df["horsepower"].quantile(0.25)
q3 = df["horsepower"].quantile(0.75)
iqr = q3-q1
ur = q3+1.5*iqr
lr = q1 - 1.5*iqr
df = df.loc[(df["horsepower"]>lr) & (df["horsepower"]<ur)]
```

```python
sns.boxplot(data=df['horsepower'])
```
