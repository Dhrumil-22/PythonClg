```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('social_media_engagement.csv')
df = pd.DataFrame(data)
df
```

```python
df.nunique()
```

```python
fig = px.bar(df,
             x='platform',
             y='likes',
             width = 800,
             height = 500,
             title = 'Social Media Analysis',
             color = 'post_type',
             barmode = 'group',
             hover_data = ['post_day','sentiment_score'],
             pattern_shape = 'post_type',
             text_auto = True)
fig.show()
```

```python
fig = px.bar(df,
             x='platform',
             y='likes',
             width = 800,
             height = 500,
             title = 'Social Media Analysis',
             color = 'post_type',
             barmode = 'stack',
             hover_data = ['post_day','sentiment_score'],
             pattern_shape = 'post_type',
             text_auto = True)
fig.show()
```

# Histogram

```python
data = pd.read_csv('social_media_engagement.csv')
df = pd.DataFrame(data)
df
```

```python
fig = px.histogram(df,
                   x = 'likes',
                   color='platform',
                   text_auto=True,
                   color_discrete_map={
                       'Facebook':'Yellow',
                       'Instagram':'skyblue',
                       'Twitter' : 'red'
                   },
                   histfunc='count',
                   color_discrete_sequence=['yellow','skublue','red'],
                   
                   )
fig.show()
```

```python
data = {
    'Marks' : [40,45,50,55,60,65,70,75,80,85],
    'Gender' : ['M','F','M','F','M','F','M','F','M','F']
}
df = pd.DataFrame(data)
fig = px.histogram(df,
                   x = 'Marks',
                   nbins=4,
                   color='Gender',
                   pattern_shape='Gender',
                   text_auto=True)
fig.show()
```

# pie chart


```python
data = {
    'Month':['Jan','Feb','Mar','Apr','Jan','Feb','Mar','Apr'],
    'Sales':[100,120,140,160,90,110,130,150],
    'Region' : ['North','North','North','North','South','South','South','South']   
}
df = pd.DataFrame(data)
df
```

```python
fig = px.pie(df,
             values='Sales',
             names='Month',
             color_discrete_sequence=['blue','green','orange','red'],
             )
fig.update_traces(textinfo = 'percent+label',
                  textposition = 'inside',
                  pull = [0,0,0.1,0.1])
fig.show()
```
