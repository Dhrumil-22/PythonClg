```python
import pandas as pd
import numpy as np
data = pd.read_csv("production_analysis_dataset.csv")
df = pd.DataFrame(data)
df
```

# Q1. Find the total units produced by each product_category.



```python
import pandas as pd
import numpy as np

df.groupby("product_category").sum(numeric_only=True)['units_produced']
```

# Q2. Calculate the average profit for each factory.


```python
df.groupby("factory").mean(numeric_only=True)["profit"]
```

# Q3. Find the total defective_units for each shift.


```python
df.groupby('shift').sum()['defective_units']
```

# Q4. Calculate the average production_time_hr for each supervisor.


```python
df.groupby('supervisor').mean(numeric_only=True)['production_time_hr']
```

```python
df.groupby()
```

# Q5. Find the total energy_consumption_kwh for each factory and shift.


```python
df.groupby(["factory",'shift']).sum(numeric_only=True)['energy_consumption_kwh']
```

# Q6. For each product_category, calculate:
total units_produced
average profit
maximum defective_units.


```python
df.groupby('product_category').agg({
    'units_produced' : 'sum',
    'profit' : 'mean',
    'defective_units' : 'max',
})

```

# Q7. For each factory, calculate:
mean production_time_hr
median machine_downtime_hr.


```python
df.groupby('factory').agg(
    {
        'production_time_hr':'mean',
        'machine_downtime_hr':'median'
    }
)
```

# Q8. For each shift, compute:
minimum profit
maximum profit
average units_produced.


```python
df.groupby('shift').agg(
    {
        'profit' : ['min','max'],
        'units_produced':'mean'
    }
)
```

# Q9. For each supervisor, calculate:
total units_produced
total defective_units.


```python
df.groupby("supervisor").agg(
    {
        'units_produced':'sum',
        'defective_units':'sum'
    }
)
```

# Q10. Using apply(), calculate the defect rate for each row:
defect_rate=defective_units/units_produced


```python
df['defect_rate'] = df.apply(lambda x : x['defective_units']/x['units_produced'],axis = 1)
df
```

# Q11. Using groupby().apply(), find the profit range (max − min) for each factory.


```python
df.groupby(['factory']).apply(lambda x : x['profit'].max()-x['profit'].min())
```

# Q12. Using groupby().apply(), find the row with maximum units_produced in each product_category.


```python
df.groupby(['product_category']).apply(lambda x : x['units_produced'].idxmax())
```

# Q13. Use apply() to create a new column efficiency_score:
efficiency=units_produced/production_time_hr


```python
df['efficiency_score'] = df.apply(lambda x : x['units_produced']/x['production_time_hr'],axis=1)
df
```
