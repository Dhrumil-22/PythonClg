```python
from bs4 import BeautifulSoup 
import pprint
import pandas as pd
```

```python
f = open("indeed jobs.html")
data = BeautifulSoup(f,'html.parser')
```

```python
jobs = data.find_all("div",class_="job_seen_beacon")
print(len(jobs))
```

```python
dict = {'Title':[],'Compny':[],'Location':[],'Discription':[]}
for i in jobs:
    t = i.find("h2",class_="jobTitle css-1psdjh5 eu4oa1w0")
    dict["Title"].append(t.text)
    
    c = i.find("div",class_="css-1afmp4o e37uo190")
    dict["Compny"].append(c.text)
    
    l = i.find("div",class_="css-1restlb eu4oa1w0")
    dict["Location"].append(l.text)
    
    d = i.find("div",class_="jobMetaDataGroup css-qspwa8 eu4oa1w0")
    # dict["Discription"].append(d.text)
    if d != None:
        dict["Discription"].append(d.text)
    else:
        dict["Discription"].append("")
        
        
    
pprint.pprint(dict)

df = pd.DataFrame(dict)
df

```
