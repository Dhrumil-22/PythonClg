```python

url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_Key}"
res = requests.get(url)
print(res)
result = res.json()
a = json.dumps(result,indent=4)
print(a)
print(len(result['list']))
```

```python
d = {'Description':[],
     'Max Temp':[],
     'Humidity':[],
     'Wind Speed':[],
     'Date_time':[]}

for i in result['list']:
    d['Description'].append(i['weather'][0]['description'])
    d['Max Temp'].append(i['main']['temp_max'])
    d['Humidity'].append(i['main']['humidity'])
    d['Wind Speed'].append(i['wind']['speed'])
    d['Date_time'].append(i['dt_txt'])
pprint.pprint(d)

    
```
