```python
import pandas as pd
from bs4 import BeautifulSoup 
import pprint
```

```python
f = open("Scrape_this.html")
data = BeautifulSoup(f,'html.parser')
```

```python
col = data.find_all("div",class_="right-part")
print(len(col))
```

```python
dict = {"company":[],"Price":[],"low":[],"high":[],"value":[]}

c = data.find_all("td",class_="")
for i in c:
    com = i.find('a')
    dict["company"].append(com.text) 


d = data.find_all("td",class_="txt-right")
cp = d[0::5]
l = d[1::5]
h = d[2::5]
vol = d[3::5]

for i in cp:
    dict["Price"].append(i.text)
for i in l:
    dict["low"].append(i.text)
for i in h:
    dict["high"].append(i.text)
for i in vol:
    dict["value"].append(i.text)
    
    
pprint.pprint(dict)
```

```python
df = pd.DataFrame(dict)
df
```
